## Suppose, several months has past. But today we need to take a look at our old project.

```python
import ipykee
```

## We need keeper who can help us to find it

```python
keeper = ipykee.Keeper("TestProject")
```

## If we forgot notebook name:

```python
keeper.list_notebooks()
```

## And keeper will give us our precious notebook!

```python
session = keeper["13-howto-ipykee-advanced"]
```

## Let's take a look at history

```python
session.history()
```

## We have only one commit because we made one commit, do you remember?  Let's take a look at results.

```python
variables = session.get_variables("master")
```

## You could replace "master" by "master~1" or by commit id. Remember, git inside.

```python
variables.keys()
```

## Mm, looks like we have report from that old notebook.

```python
variables["report"].features_correlation_matrix().plot()
```

## It works!

## And we even can use saved classifier

```python
classifier = variables["classifier"]
```

```python
classifier.features
```

```python
classifier.get_learning_curve().plot()
```
