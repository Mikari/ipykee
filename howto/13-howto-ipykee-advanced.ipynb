```python
# just some stuff to prepare data and variables
from cern_utils import converter_csv
variables = ["FlightDistance", "FlightDistanceError", "IP", "IPSig", "VertexChi2", "pt", "p0_IPSig", "p1_IPSig", "p2_IPSig", "p0_IP",
             "p1_IP", "p2_IP", 'p0_Laura_IsoBDT', 'p1_Laura_IsoBDT', 'p2_Laura_IsoBDT', "DOCAone", "DOCAtwo", "DOCAthree", "isolationd",
             "isolatione", "isolationf", "p0_pt", "p1_pt", "p2_pt", "CDF3", "CDF1", "CDF2", "Laura_SumBDT"
            ]
```

```python
signal_file = converter_csv.load_from_csv(
    'ToyMC/toyMC_sig_mass.csv', 
    sep='\t')
bck_file = converter_csv.load_from_csv(
    'ToyMC/toyMC_bck_mass.csv', 
    sep='\t')
```

```python
signal_train, signal_test = signal_file.get_train_test(train_size=0.5, test_size=0.5)
bck_train, bck_test = bck_file.get_train_test(train_size=0.5, test_size=0.5)
```

```python
from cern_utils import ef_classifier
# configuring classifier
ef = ef_classifier.ClassifierEF(dataset_name='test-project', formula_name='some-formula-from-howto-ipykee-advanced-20141010', 
                                directory='some-path', url_base='http://efilter.cern.ch/', user='lhcb', password='yandex')
ef.set_params(features=variables, iterations = 100, spectators=['mass'], sync=True)

# training classifier
ef.fit(signal_train, bck_train)
```

```python
# get prediction on data after classification
from cern_utils.predictions_report import PredictionsInfo
report = PredictionsInfo({"classifier": ef}, signal_test, bck_test)
```

```python
report.prediction_pdf().plot()
```

```python
report.features_correlation_matrix().plot()
```
